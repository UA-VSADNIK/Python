## **3. Пам’ятки культури та природи**
### **Культурні пам'ятки Ісландії:**
1. *<a href="https://uk.wikipedia.org/wiki/%D0%93%D0%B0%D1%82%D0%BB%D1%8C%D2%91%D1%80%D1%96%D0%BC%D1%81%D0%BA%D1%96%D1%80%D0%BA%27%D1%8F">Церква Гатльгрімскіркья.</a>* Побудована у формі вулканічних базальтових колон, ця церква є найбільш впізнаваною будівлею в країні. З вежі відкривається чудовий краєвид на столицю.

2. *<a href="https://navicup.com/object/iceland-grand-tour/arbaer-open-air-museum-308064/ua">Музей Арбаер в Рейк'явіку.</a>* Історичний музей просто неба, де відтворюється традиційне ісландське село з хатинами, критими дерновими дахами.

3. *<a href="https://ru.wikipedia.org/wiki/%D0%A5%D0%B0%D1%80%D0%BF%D0%B0">Харпа.</a>* Вражаючий концертний та конференц-центр у Рейк'явіку. Будівля отримала кілька архітектурних нагород за свій сучасний стиль і вражаючий вигляд, особливо у вечірній час.

### **Природні пам'ятки Ісландії:**
*<a href="https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%86%D1%96%D0%BE%D0%BD%D0%B0%D0%BB%D1%8C%D0%BD%D0%B8%D0%B9_%D0%BF%D0%B0%D1%80%D0%BA_%D0%92%D0%B0%D1%82%D0%BD%D0%B0%D0%B9%D0%BE%D0%BA%D1%83%D1%82%D0%BB%D1%8C#:~:text=%D0%9D%D0%B0%D1%86%D1%96%D0%BE%D0%BD%D0%B0%D0%BB%D1%8C%D0%BD%D0%B8%D0%B9%20%D0%BF%D0%B0%D1%80%D0%BA%20%D0%92%D0%B0%D1%82%D0%BD%D0%B0%D0%B9%D0%BE%D0%BA%D1%83%D1%82%D0%BB%D1%8C%20(%D1%96%D1%81%D0%BB.,%D1%96%D1%81%D0%BD%D1%83%D0%B2%D0%B0%D0%BB%D0%B8%20%D1%82%D1%83%D1%82%20%D1%80%D0%B0%D0%BD%D1%96%D1%88%D0%B5%2C%20%D0%BD%D0%B0%D0%BF%D1%80%D0%B8%D0%BA%D0%BB%D0%B0%D0%B4%20%D0%A1%D0%BA%D0%B0%D0%B2%D1%82%D0%B0%D1%84%D0%B5%D1%82%D0%BB%D1%8C.">Національний парк Ватнайокудль.</a>* Один з найбільших національних парків Європи, який охоплює однойменний льодовик Ватнайокудль — найбільший у Європі за обсягом. 2. *<a href="https://uk.wikipedia.org/wiki/%D0%94%D0%B5%D1%82%D1%82%D1%96%D1%84%D0%BE%D1%81%D1%81">Водоспад Деттіфосс.</a>* Відомий як наймогутніший водоспад Європи, його висота — 44 метри. Вражаюча міць води створює особливу атмосферу цього місця.

3. *<a href="https://uk.wikipedia.org/wiki/%D0%93%D0%B5%D0%BA%D0%BB%D0%B0">Вулкан Гекла.</a>* Один з найактивніших вулканів у світі. Впродовж історії Гекла вивергалася більше 20 разів, що робить її важливим геологічним об’єктом.

5. *<a href="https://uk.wikipedia.org/wiki/%D2%90%D1%83%D0%BB%D1%8C%D1%84%D0%BE%D1%81%D1%81">Водоспад Ґуллфосс.</a>* Один з найвідоміших водоспадів Ісландії, який знаходиться на річці Хвітау. Дві ступені водоспаду створюють надзвичайно ефектний вигляд.
